### Training a Snake Agent using DQN

```python
from conductor import Conductor

c = Conductor()
scores = c.train(episodes=3000,
        gamma=0.99,
        epsilon_start=0.2,
        epsilon_decay=0.995,
        batch_size=64,
        target_update_freq=1000)
```

```python
# plot steps per episode

import matplotlib.pyplot as plt

plt.plot(list(range(len(scores))), scores)
plt.title("Score per episode")
plt.xlabel("Episode")
plt.ylabel("Score")
plt.show()
```
