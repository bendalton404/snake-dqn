### Training a Snake Agent using DQN

```python
from conductor import Conductor

c = Conductor()
scores = c.train(episodes=3000,
        gamma=0.99,
        epsilon_start=1.0,
        epsilon_end=0.03,
        epsilon_decay=0.997,
        batch_size=32,
        target_update_steps=500)
```

```python
# plot steps per episode

import matplotlib.pyplot as plt

plt.plot(list(range(len(scores))), scores)
plt.title("Score per episode")
plt.xlabel("Episode")
plt.ylabel("Score")
plt.show()
```
